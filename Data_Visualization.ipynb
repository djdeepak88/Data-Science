```python
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x), color='yellow')
```

```python
x = np.linspace(0,10,100)
#Background grid style.
plt.style.use('seaborn-whitegrid')
plt.plot(x,np.sin(x), color='yellow')
plt.plot(x,np.sin(x-1), color='green')
plt.plot(x,np.sin(x-2), color='blue')
plt.plot(x,np.sin(x-3), color='black')
```

```python
plt.show()
```

```python

plt.plot(x,x+1, color='yellow')
plt.plot(x,x+2, color='green')
plt.plot(x,x+3, color='blue')
```

```python
plt.show()
```

```python
plt.plot(x,np.sin(x), label='sin')
plt.plot(x,np.cos(x), label='cos')
plt.xlim(-5,11)
plt.ylim(-1.5,1.5)
plt.xlabel("x-axis")
plt.ylabel("sinx")
plt.legend()
```

```python
plt.show()
```

```python
plt.plot(x,np.tan(x),label='tan')
plt.xlim(-5,10)
plt.ylim(-5,5)
plt.xlabel("X-Axis")
plt.ylabel("Tan(x)")
plt.legend()
```

```python
plt.show()
```

```python
#Generates Random Number generator Object
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100) # generates + & - numbers
```

```python
colors = rng.rand(100) # generates positive numbers
sizes = 1000* rng.rand(100)
```

```python
#alpha is transparency level.
# 4 types of data is plotted here.
plt.scatter(x,y,c=colors,s=sizes, alpha=0.3, cmap='viridis')
```

```python
plt.show()
```

```python
plt.scatter(x,y,c=colors,s=sizes, alpha=0.3, cmap='magma')
```

```python
plt.show()
```

```python
plt.scatter(x,y,c=colors,s=sizes, alpha=0.3, cmap='plasma')
```

```python
plt.show()
```

```python
plt.scatter(x,y,c=colors,s=sizes, alpha=0.3, cmap='inferno')
```

```python
plt.show()
```
